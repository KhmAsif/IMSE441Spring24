```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import hypergeom, poisson,expon
```

```python
#Q4-A
A = hypergeom.pmf(k=2,M=500,n=50,N=20)
print('Q4-A = ', A)
#B
B = hypergeom.cdf(k=1,M=500,n=50,N=20)
print('Q4-B = ', B)
#C - is k 2 or 3
C = hypergeom.sf(k=2,M=500,n=50,N=20)
print('Q4-C = ', C)
#D
mean, var = hypergeom.stats(M=500,n=50,N=20)
print('Q4-D Mean = ', mean, 'Q4-D Variance = ', var)
```

```python
#E
x = np.arange(0, 40)
mu=1.5*10
pmf = poisson.pmf(x, mu)


plt.vlines(x,0, pmf, colors='b', lw=5, alpha=0.5)

plt.xlabel('Time (minutes)')
plt.ylabel('Probability')
plt.title('Poisson Distribution (mu = 1.5 *10)')

plt.show()
```
