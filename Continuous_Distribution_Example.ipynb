**Weibull Distribution**

```python
from scipy.stats.distributions import weibull_min
```

```python
#Finding the probalibty of bearing lifetime more that 6000 hours, shape=2, scale=5000
#weibull_min=(x,c=shape=beta,scale=delta=1) scale default value is 1 for python
P6000=1-weibull_min.cdf(6000,c=2,scale=5000)
P6000
```

```python
p5000=weibull_min.cdf(5000,c=2,scale=5000)
p5000
```

```python
#Finding mean, Variance and Standard Deviation
#weibull_min.mean(c=shape parameter, scale= scale parameter)
m= weibull_min.mean(c=2,scale=5000)
m
```

```python
v= weibull_min.var(c=2,scale=5000)
v
```

```python
s=weibull_min.std(c=2,scale=5000)
s
```

```python
mean, var = weibull_min.stats(c=2, scale=500)
print('mean is', mean, 'variance is', var)
```
