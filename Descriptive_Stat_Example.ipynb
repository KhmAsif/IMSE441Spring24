```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data upload
c_str= pd.read_csv('/content/drive/MyDrive/IMSE_441/compressive_strength.csv')
c_str.head()
```

```python
grp_str=pd.read_csv('/content/drive/MyDrive/IMSE_441/Grip_strength.csv')
grp_str.head()
```

```python
c_str.describe()
```

```python
grp_str.describe()
```

```python
#Mean, Median, Mode
str_mean=grp_str['Grip_Strength'].mean()
str_median=grp_str['Grip_Strength'].median()
str_mode=grp_str['Grip_Strength'].mode()

print('Mean Strength= ',str_mean,'Median Strength= ',str_median, 'Mode Strength=',str_mode)
```

```python
#Range
Range= grp_str['Grip_Strength'].max()-grp_str['Grip_Strength'].min()
Range
```

```python
#IQR
Q1= np.percentile(c_str['cs'],(25))
Q2= np.percentile(c_str['cs'],(50))
Q3= np.percentile(c_str['cs'],(75))
print('1st quartile = ',Q1, '2nd quartile = ',Q2, '3rd quartile = ',Q3)
```

```python
#Variance
c_str['cs'].var()
```

```python
#Standard Deviation
c_str['cs'].std()
```

```python
#Time series
plt.figure(figsize=(10,5))
plt.plot(c_str['cs'],linewidth=3, marker= 'o')
plt.title('Time Series' )
plt.xlabel('sequence')
plt.ylabel('compressive strength')
plt.show()
```

```python
#Histogram
plt.hist(c_str['cs'], bins=9, rwidth=0.95, color='green', alpha=0.75)
plt.xlabel('Strength')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(alpha=0.25)
plt.show()
```

```python
plt.hist(c_str['cs'], cumulative= 'True', bins=9, rwidth=0.95, color='green', alpha=0.75)
plt.xlabel('Strength')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(alpha=0.25)
plt.show()
```

```python
#Boxplot
plt.boxplot([grp_str['Shoe_size'],grp_str['Height']], labels=['Shoe_size', 'Height'])
plt.xlabel('Data')
plt.ylabel('YData')
plt.xlim(0,3)
plt.ylim(4,15)
plt.title('Boxplot')
plt.show()
```
